## Descrizione del dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
dataframe = pd.read_csv('main_dataframe.csv', index_col=0)
dataframe.columns = pd.MultiIndex.from_tuples([eval(a) for a in dataframe.columns])
dataframe
```

```python
#
# Calcoliamo la media di tutti i valori per ogni attività
#

df = dataframe.drop([('p','',''), ('s','','')], axis=1)
df = df.groupby([('a','','')]).mean()
df = df.rename_axis(['a'])
df
```

```python
df['T','xacc','std'].plot(kind='bar')
plt.figure()
df['T','yacc','std'].plot(kind='bar')
plt.figure()
df['T','zacc','std'].plot(kind='bar')
```

Prendendo come riferimento i dati di accelerazione del Torso, notiamo che i valori della deviazione standard per le varie attività sono simili per tutti e 3 gli assi (x, y, z), anche se su scale diverse.

```python
#
# Computiamo la deviazione standard media dei tre assi per ogni sensore accelerometro utilizzato
#

acc_standard_dev = pd.DataFrame()
acc_standard_dev['T'] = (df[('T', 'xacc', 'std')] + df[('T', 'yacc', 'std')] + df[('T', 'zacc', 'std')]) / 3
acc_standard_dev['LA'] = (df[('LA', 'xacc', 'std')] + df[('LA', 'yacc', 'std')] + df[('LA', 'zacc', 'std')]) / 3
acc_standard_dev['RA'] = (df[('RA', 'xacc', 'std')] + df[('RA', 'yacc', 'std')] + df[('RA', 'zacc', 'std')]) / 3
acc_standard_dev['LL'] = (df[('LL', 'xacc', 'std')] + df[('LL', 'yacc', 'std')] + df[('LL', 'zacc', 'std')]) / 3
acc_standard_dev['RL'] = (df[('RL', 'xacc', 'std')] + df[('RL', 'yacc', 'std')] + df[('RL', 'zacc', 'std')]) / 3
acc_standard_dev
```

```python
plt.title('Torso')
acc_standard_dev['T'].plot(kind='bar')
plt.figure()
plt.title('Left arm')
acc_standard_dev['LA'].plot(kind='bar')
plt.figure()
plt.title('Right arm')
acc_standard_dev['RA'].plot(kind='bar')
plt.figure()
plt.title('Left leg')
acc_standard_dev['LL'].plot(kind='bar')
plt.figure()
plt.title('Rigth leg')
acc_standard_dev['RL'].plot(kind='bar')
```

In più o meno tutte le attività la gamba sinistra si muove quanto la gamba destra, così come braccio sinistro e destro. Inoltre in media, nelle varie attività, le gambe si muovono più delle braccia.
Nelle attività 15 e 16 (cyclette in orizzontale e verticale) notiamo un valore della deviazione standard maggiore nelle gambe rispetto alle braccia, a differenza dell'attività 19 (basket) in cui sono le braccia a fare movimenti più complessi.  
Possiamo notare infine che il valore maggiore per le gambe si ottiene nell'attività 12 (corsa intensa), mentre per le braccia è l'attività 19 (basket).

```python
#
# Computiamo la deviazione standard media dei tre assi per ogni sensore giroscopio utilizzato
#

gyro_standard_dev = pd.DataFrame()
gyro_standard_dev['T'] = (df[('T', 'xgyro', 'std')] + df[('T', 'ygyro', 'std')] + df[('T', 'zgyro', 'std')]) / 3
gyro_standard_dev['LA'] = (df[('LA', 'xgyro', 'std')] + df[('LA', 'ygyro', 'std')] + df[('LA', 'zgyro', 'std')]) / 3
gyro_standard_dev['RA'] = (df[('RA', 'xgyro', 'std')] + df[('RA', 'ygyro', 'std')] + df[('RA', 'zgyro', 'std')]) / 3
gyro_standard_dev['LL'] = (df[('LL', 'xgyro', 'std')] + df[('LL', 'ygyro', 'std')] + df[('LL', 'zgyro', 'std')]) / 3
gyro_standard_dev['RL'] = (df[('RL', 'xgyro', 'std')] + df[('RL', 'ygyro', 'std')] + df[('RL', 'zgyro', 'std')]) / 3
gyro_standard_dev
```

```python
plt.title('Torso')
gyro_standard_dev['T'].plot(kind='bar')
plt.figure()
plt.title('Left arm')
gyro_standard_dev['LA'].plot(kind='bar')
plt.figure()
plt.title('Right arm')
gyro_standard_dev['RA'].plot(kind='bar')
plt.figure()
plt.title('Left leg')
gyro_standard_dev['LL'].plot(kind='bar')
plt.figure()
plt.title('Rigth leg')
gyro_standard_dev['RL'].plot(kind='bar')
```

```python
#
# Computiamo la deviazione standard media dei tre assi per ogni sensore magnetometro utilizzato
#

mag_standard_dev = pd.DataFrame()
mag_standard_dev['T'] = (df[('T', 'xmag', 'std')] + df[('T', 'ymag', 'std')] + df[('T', 'zmag', 'std')]) / 3
mag_standard_dev['LA'] = (df[('LA', 'xmag', 'std')] + df[('LA', 'ymag', 'std')] + df[('LA', 'zmag', 'std')]) / 3
mag_standard_dev['RA'] = (df[('RA', 'xmag', 'std')] + df[('RA', 'ymag', 'std')] + df[('RA', 'zmag', 'std')]) / 3
mag_standard_dev['LL'] = (df[('LL', 'xmag', 'std')] + df[('LL', 'ymag', 'std')] + df[('LL', 'zmag', 'std')]) / 3
mag_standard_dev['RL'] = (df[('RL', 'xmag', 'std')] + df[('RL', 'ymag', 'std')] + df[('RL', 'zmag', 'std')]) / 3
mag_standard_dev
```

```python
plt.title('Torso')
gyro_standard_dev['T'].plot(kind='bar')
plt.figure()
plt.title('Left arm')
gyro_standard_dev['LA'].plot(kind='bar')
plt.figure()
plt.title('Right arm')
gyro_standard_dev['RA'].plot(kind='bar')
plt.figure()
plt.title('Left leg')
gyro_standard_dev['LL'].plot(kind='bar')
plt.figure()
plt.title('Rigth leg')
gyro_standard_dev['RL'].plot(kind='bar')
```

La distribuzione dei valori della deviazione standard per gli altri sensori, giroscopio e magnetometro, è molto simile a quella dell'accelerometro. Dunque valgono gli stessi ragionamenti esposti in precedenza.
