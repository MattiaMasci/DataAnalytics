## Descrizione del dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#
# Creiamo un dataframe con tutti i dati del dataset
#

dataframe_all = pd.read_csv('main_dataframe_all.csv')
dataframe_all.columns = pd.MultiIndex.from_tuples([eval(a) for a in dataframe_all.columns])

dataframe_all = dataframe_all.groupby([('a',''), ('p',''), ('s',''), ('idx','')]).mean()
dataframe_all = dataframe_all.rename_axis(['a', 'p', 's', 'i'])
dataframe_all
```

```python
#
# Sitting - Torso - Gyroscope
#

dataframe_all.loc[(1, 1)]['T','xgyro'].plot()

#
# NB: funzione approssimativamente costante
#
```

```python
#
# Standing in elevator - Torso - Accelerometer
#

dataframe_all.loc[(7, 8)]['T','xacc'].plot()

#
# NB: quando l'ascensore si muove/si ferma c'è una piccola accelerazione/decelerazione
#
```

```python
#
# Walk - Left/Right arm - Accelerometer
#

dataframe_all.loc[(9, 1, 11)]['RA','xacc'].plot()
dataframe_all.loc[(9, 1, 11)]['LA','xacc'].plot()

#
# NB: funzione periodica; i movimenti delle braccia sono coerenti
#
```

```python
#
# Walk/Run - Right Leg - Accelerometer
#

dataframe_all.loc[(10, 1, 11)]['RL','yacc'].plot()
dataframe_all.loc[(12, 1, 11)]['RL','yacc'].plot()

#
# NB: differenza tra camminata e corsa
#     i movimenti sono simili ma nella corsa le accelerazioni sono più intense
#
```

```python
#
# Stepper - Torso - Accelerometer
#

for i in range(1,4):
    dataframe_all.loc[(13, i, 11)]['T','xacc'].plot()

#
# NB: stesso allenamento; persone diverse
#     i dati sono molto simili per vari i componenti
#
```

```python
#
# Rowing (canoa) - Arms - Accelerometer
#

dataframe_all.loc[(17, 6, slice(10, 15))]['T','xacc'].plot()

#
# NB: movimento periodico, sebbene si nota un periodo più lento rispetto ai precedenti (es. camminata o corsa)
#
```

```python
#
# Jump - Torso - Accelerometer
#

dataframe_all.loc[(18, 1, 10)]['T','xacc'].plot()

#
# NB: funzione periodica, con picchi di accelerazione molto alti
#
```

```python
#
# Basket - Torso - Acc/Gyro/Mag
#

dataframe_all.loc[(19, 2)]['T','xacc'].plot()
plt.figure()
dataframe_all.loc[(19, 2)]['T','xgyro'].plot()
plt.figure()
dataframe_all.loc[(19, 2)]['T','ymag'].plot()

#
# NB: molto caotico poiché i movimenti sono molto irregolari
#
```

## Correlation Matrix

```python
import pandas as pd
import numpy as np
# import seaborn as sn
import matplotlib.pyplot as plt

dataframe = pd.read_csv('clean_dataframe.csv', index_col=0)
dataframe.columns = pd.MultiIndex.from_tuples([eval(a) for a in dataframe.columns])
dataframe
```

```python
corrMatrix = dataframe.corr()
corrMatrix
```

```python
#
# Valori mostrati nelle etichette degli gli assi
#

names = dict( enumerate(corrMatrix.index) )
ticks = np.arange(315, step=63)
labels = [names[i] for i in ticks]


fig, ax = plt.subplots(figsize=(30, 30))

plt.tick_params(labelsize=20)
plt.xticks(rotation='vertical')
ax.set_xticks(ticks)
ax.set_xticklabels(labels)
ax.set_yticks(ticks)
ax.set_yticklabels(labels)

im = ax.imshow(corrMatrix, interpolation='nearest', aspect='auto', cmap='coolwarm')
fig.colorbar(im)

plt.show()
```

Non c'è molta correlazione tra le varie colonne, né positiva né negativa.

Possiamo notare bassa correlazione tra i movimenti delle due gambe, e ancor di meno tra quelli delle due braccia.  
Infine sulla diagonale principale si creano spesso dei piccoli 'quadrati' che mostrano una correlazione tra i 3 assi (x, y, z) nello stesso sensore (accelerometro, giroscopio, magnetometro).

Tra questi 3 sensori, il magnetometro è quello con minore correlazione rispetto agli altri, creando la griglia di linee bianche.
